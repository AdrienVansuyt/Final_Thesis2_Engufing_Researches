# **Les Patterns Englobantes (Engulfing) et Script de Screener Multi-Unité de Temps**

# **Introduction**

Les patterns englobantes, aussi appelées "engulfing patterns" en anglais, sont des configurations de chandeliers japonais qui signalent souvent des retournements de tendance. Elles sont considérées comme des indicateurs puissants dans l'analyse technique et sont utilisées par de nombreux traders pour identifier des opportunités d'achat ou de vente. Ce cours vise à expliquer ce que sont les patterns englobantes, comment les repérer, et comment les utiliser dans vos stratégies de trading.

# **Les Chandeliers Japonais**

Avant de plonger dans les patterns englobantes, il est crucial de comprendre les bases des chandeliers japonais. Un chandelier se compose de quatre points principaux :

- **Ouverture (Open) :** Le prix auquel un actif commence à se négocier lors d'une période spécifique.
- **Fermeture (Close) :** Le prix auquel un actif finit de se négocier à la fin de la période.

- **Haut (High) :** Le prix le plus élevé atteint par l'actif pendant la période.

- **Bas (Low) :** Le prix le plus bas atteint par l'actif pendant la période.
Les chandeliers sont souvent colorés pour indiquer la direction du mouvement de prix :

**Vert ou blanc :** Le prix a augmenté (fermeture > ouverture).

**Rouge ou noir :** Le prix a diminué (fermeture < ouverture).
Les Patterns Englobantes

**Il existe deux types principaux de patterns englobantes :**

- Englobante Haussière (Bullish Engulfing) : Indique un potentiel retournement à la hausse.
- Englobante Baissière (Bearish Engulfing) : Indique un potentiel retournement à la baisse.


**Une englobante haussière se forme lorsque :**

Un petit chandelier baissier (rouge) est suivi d'un grand chandelier haussier (vert).
Le corps du chandelier haussier englobe entièrement le corps du chandelier baissier précédent.
Cela signifie que les acheteurs ont pris le contrôle du marché, poussant le prix au-delà de l'ouverture et de la fermeture du chandelier précédent.


**Une englobante baissière se forme lorsque :**

Un petit chandelier haussier (vert) est suivi d'un grand chandelier baissier (rouge).
Le corps du chandelier baissier englobe entièrement le corps du chandelier haussier précédent.
Cela signifie que les vendeurs ont pris le contrôle du marché, faisant chuter le prix au-delà de l'ouverture et de la fermeture du chandelier précédent.

**Comment Utiliser les Patterns Englobantes**

Les patterns englobantes peuvent être utilisées dans divers cadres de temps, du court terme (comme les graphiques en minute) au long terme (comme les graphiques mensuels). Voici quelques étapes pour les utiliser efficacement :

- **Identification :** Repérez les patterns englobantes sur vos graphiques en utilisant les critères mentionnés ci-dessus.

- **Confirmation :** Utilisez d'autres indicateurs techniques pour confirmer le signal donné par la pattern englobante. Par exemple, un indicateur de volume peut montrer une augmentation du volume lors de la formation de la pattern, ce qui renforce le signal.
- **Entrée en Position :** Après confirmation, prenez une position en suivant le signal :
  - Pour une englobante haussière, considérez une position longue (achat).
  - Pour une englobante baissière, considérez une position courte (vente).

- **Gestion des Risques :** Placez des stops appropriés pour gérer votre risque. Par exemple, pour une position longue sur une englobante haussière, placez un stop-loss en dessous du plus bas du chandelier haussier.
Conclusion

Les patterns englobantes sont des outils précieux pour les traders cherchant à identifier des retournements potentiels du marché. Bien qu'elles ne garantissent pas toujours un retournement, leur utilisation combinée avec d'autres indicateurs techniques peut améliorer considérablement vos chances de succès en trading. Commencez à pratiquer l'identification et l'utilisation de ces patterns sur vos graphiques et intégrez-les dans votre stratégie de trading pour voir comment elles peuvent vous aider à prendre des décisions plus informées.

# **Exemple de Script d'Analyse des Patterns Englobantes**

Voici un script Python pour analyser les patterns englobantes sur plusieurs actifs et unités de temps. Le script télécharge les données financières, détecte les patterns englobantes, et affiche des heatmaps des fréquences et des mouvements moyens après les patterns.

Objectifs du Script

# Notre script a pour but de : **bold text**

1. Télécharger les données financières pour plusieurs actifs et unités de temps.
2. Détecter les patterns englobantes haussières et baissières.
3. Analyser les mouvements de prix après l'apparition de ces patterns.
4. Calculer la fréquence des patterns englobantes.
5. Faire un score fréquence du pattern x pips potentiels
6. Visualiser les résultats sous forme de heatmaps et graphiques.

```python
import yfinance as yf
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Listes des symboles par catégorie
actions = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA', 'META']
indices = ['^GSPC', '^IXIC', '^DJI', '^GDAXI', '^FTSE', '^N225', '^FCHI']
forex = ['EURUSD=X', 'JPY=X', 'GBPUSD=X', 'AUDUSD=X', 'NZDUSD=X', 'USDCAD=X', 'USDCHF=X']
commodities = ['CL=F', 'GC=F', 'SI=F', 'HG=F', 'NG=F']
cryptos = ['BTC-USD', 'ETH-USD', 'BNB-USD', 'USDT-USD', 'ADA-USD']
obligations = ['^TNX', '^TYX']

# Unités de temps (du plus petit au plus grand)
timeframes = ['30m', '60m', '90m', '1h', '1d', '5d', '1wk', '1mo', '3mo']

# Mapper les unités de temps avec des périodes compatibles
period_map = {
    '30m': '1mo',
    '60m': '1mo',
    '90m': '1mo',
    '1h': '1mo',
    '1d': '3mo',
    '5d': '6mo',
    '1wk': '1y',
    '1mo': '2y',
    '3mo': '5y'
}

# Fonction pour détecter les patterns englobantes et mesurer les mouvements et points culminants/plus bas
def detect_engulfing(data, long_term_horizon):
    signals = {'bullish_engulfing': [], 'bearish_engulfing': []}
    movements = {'bullish': [], 'bearish': []}
    long_term_movements = {'bullish': [], 'bearish': []}
    for i in range(1, len(data)):
        # Bullish Engulfing
        if data['Open'][i] < data['Close'][i-1] and data['Close'][i] > data['Open'][i-1]:
            signals['bullish_engulfing'].append(i)
            # Calculer le mouvement suivant l'englobante haussière jusqu'à un chandelier baissier
            for j in range(i+1, len(data)):
                if data['Close'][j] < data['Open'][j]:
                    movements['bullish'].append(data['Close'][j] - data['Close'][i])
                    break
            # Calculer le point culminant sur l'horizon de temps long
            if i + long_term_horizon < len(data):
                long_term_movements['bullish'].append(max(data['Close'][i:i+long_term_horizon]) - data['Close'][i])

        # Bearish Engulfing
        elif data['Open'][i] > data['Close'][i-1] and data['Close'][i] < data['Open'][i-1]:
            signals['bearish_engulfing'].append(i)
            # Calculer le mouvement suivant l'englobante baissière jusqu'à un chandelier haussier
            for j in range(i+1, len(data)):
                if data['Close'][j] > data['Open'][j]:
                    movements['bearish'].append(data['Close'][j] - data['Close'][i])
                    break
            # Calculer le point le plus bas sur l'horizon de temps long
            if i + long_term_horizon < len(data):
                long_term_movements['bearish'].append(min(data['Close'][i:i+long_term_horizon]) - data['Close'][i])
    return signals, movements, long_term_movements

# Détecter les patterns pour chaque actif et chaque unité de temps par catégorie
categories = {'Actions': actions, 'Indices': indices, 'Forex': forex, 'Commodities': commodities, 'Cryptos': cryptos, 'Obligations': obligations}
engulfing_counts = {category: {tf: {'bullish': [], 'bearish': [], 'bullish_movements': [], 'bearish_movements': [], 'bullish_long_term': [], 'bearish_long_term': []} for tf in timeframes} for category in categories}

long_term_horizon = 20  # 20 periods for long-term analysis

for category, tickers in categories.items():
    for ticker in tickers:
        for tf in timeframes:
            try:
                period = period_map[tf]
                data = yf.download(ticker, period=period, interval=tf)
                if not data.empty:
                    signals, movements, long_term_movements = detect_engulfing(data, long_term_horizon)
                    engulfing_counts[category][tf]['bullish'].extend(signals['bullish_engulfing'])
                    engulfing_counts[category][tf]['bearish'].extend(signals['bearish_engulfing'])
                    engulfing_counts[category][tf]['bullish_movements'].extend(movements['bullish'])
                    engulfing_counts[category][tf]['bearish_movements'].extend(movements['bearish'])
                    engulfing_counts[category][tf]['bullish_long_term'].extend(long_term_movements['bullish'])
                    engulfing_counts[category][tf]['bearish_long_term'].extend(long_term_movements['bearish'])
                else:
                    print(f"No data available for {ticker} with interval {tf}")
            except Exception as e:
                print(f"Error downloading {ticker} with interval {tf}: {e}")

# Calculer la moyenne des mouvements pour chaque unité de temps et chaque catégorie
bullish_movements_mean_df = pd.DataFrame({tf: [np.mean(engulfing_counts[category][tf]['bullish_movements']) if engulfing_counts[category][tf]['bullish_movements'] else 0 for category in categories] for tf in timeframes}, index=categories)
bearish_movements_mean_df = pd.DataFrame({tf: [np.mean(engulfing_counts[category][tf]['bearish_movements']) if engulfing_counts[category][tf]['bearish_movements'] else 0 for category in categories] for tf in timeframes}, index=categories)

# Calculer la moyenne des points culminants/plus bas pour chaque unité de temps et chaque catégorie
bullish_long_term_mean_df = pd.DataFrame({tf: [np.mean(engulfing_counts[category][tf]['bullish_long_term']) if engulfing_counts[category][tf]['bullish_long_term'] else 0 for category in categories] for tf in timeframes}, index=categories)
bearish_long_term_mean_df = pd.DataFrame({tf: [np.mean(engulfing_counts[category][tf]['bearish_long_term']) if engulfing_counts[category][tf]['bearish_long_term'] else 0 for category in categories] for tf in timeframes}, index=categories)

# Calculer la fréquence des patterns englobantes par catégorie
bullish_engulfing_freq_df = pd.DataFrame({tf: [len(engulfing_counts[category][tf]['bullish']) for category in categories] for tf in timeframes}, index=categories)
bearish_engulfing_freq_df = pd.DataFrame({tf: [len(engulfing_counts[category][tf]['bearish']) for category in categories] for tf in timeframes}, index=categories)

# Calculer les scores combinés (fréquence * moyenne des mouvements)
bullish_combined_score_df = bullish_engulfing_freq_df * bullish_movements_mean_df
bearish_combined_score_df = bearish_engulfing_freq_df * bearish_movements_mean_df

# Tracer les heatmaps des fréquences des patterns englobantes par catégorie
plt.figure(figsize=(14, 7))
sns.heatmap(bullish_engulfing_freq_df, annot=True, fmt='d', cmap='Blues', linewidths=0.5, linecolor='black')
plt.title('Frequency of Bullish Engulfing Patterns by Category')
plt.xlabel('Timeframes')
plt.ylabel('Categories')
plt.show()

plt.figure(figsize=(14, 7))
sns.heatmap(bearish_engulfing_freq_df, annot=True, fmt='d', cmap='Purples', linewidths=0.5, linecolor='black')
plt.title('Frequency of Bearish Engulfing Patterns by Category')
plt.xlabel('Timeframes')
plt.ylabel('Categories')
plt.show()

# Tracer les heatmaps des mouvements moyens après englobantes (pips) par catégorie
plt.figure(figsize=(14, 7))
sns.heatmap(bullish_movements_mean_df.round(2), annot=True, fmt='.2f', cmap='Blues', linewidths=0.5, linecolor='black')
plt.title('Average Pips After Bullish Engulfing Patterns by Category')
plt.xlabel('Timeframes')
plt.ylabel('Categories')
plt.show()

plt.figure(figsize=(14, 7))
sns.heatmap(bearish_movements_mean_df.round(2), annot=True, fmt='.2f', cmap='Purples', linewidths=0.5, linecolor='black')
plt.title('Average Pips After Bearish Engulfing Patterns by Category')
plt.xlabel('Timeframes')
plt.ylabel('Categories')
plt.show()

# Tracer les heatmaps des points culminants/plus bas après englobantes (pips) par catégorie
plt.figure(figsize=(14, 7))
sns.heatmap(bullish_long_term_mean_df.round(2), annot=True, fmt='.2f', cmap='Greens', linewidths=0.5, linecolor='black')
plt.title('Average Long-Term Peak Pips After Bullish Engulfing Patterns by Category')
plt.xlabel('Timeframes')
plt.ylabel('Categories')
plt.show()

plt.figure(figsize=(14, 7))
sns.heatmap(bearish_long_term_mean_df.round(2), annot=True, fmt='.2f', cmap='Oranges', linewidths=0.5, linecolor='black')
plt.title('Average Long-Term Lowest Pips After Bearish Engulfing Patterns by Category')
plt.xlabel('Timeframes')
plt.ylabel('Categories')
plt.show()

# Tracer les heatmaps des scores combinés (fréquence * moyenne des mouvements) par catégorie
plt.figure(figsize=(14, 7))
sns.heatmap(bullish_combined_score_df.round(2), annot=True, fmt='.2f', cmap='Blues', linewidths=0.5, linecolor='black')
plt.title('Combined Score (Frequency * Average Pips) for Bullish Engulfing Patterns by Category')
plt.xlabel('Timeframes')
plt.ylabel('Categories')
plt.show()

plt.figure(figsize=(14, 7))
sns.heatmap(bearish_combined_score_df.round(2), annot=True, fmt='.2f', cmap='Purples', linewidths=0.5, linecolor='black')
plt.title('Combined Score (Frequency * Average Pips) for Bearish Engulfing Patterns by Category')
plt.xlabel('Timeframes')
plt.ylabel('Categories')
plt.show()

# Sélection des unités de temps les plus intéressantes
def select_best_timeframes(score_df):
    best_timeframes = {}
    for category in score_df.index:
        best_timeframes[category] = score_df.loc[category].idxmax()
    return best_timeframes

best_bullish_timeframes = select_best_timeframes(bullish_combined_score_df)
best_bearish_timeframes = select_best_timeframes(bearish_combined_score_df)

print("Best timeframes for Bullish Engulfing Patterns by Category:")
for category, tf in best_bullish_timeframes.items():
    print(f"{category}: {tf}")

print("\nBest timeframes for Bearish Engulfing Patterns by Category:")
for category, tf in best_bearish_timeframes.items():
    print(f"{category}: {tf}")

```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Liste des symboles
tickers = [
    'NVDA', 'CL=F', 'GC=F', 'AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA',
    'META', 'BTC-USD', 'ETH-USD', '^GSPC', '^IXIC', '^DJI', '^GDAXI', '^FTSE', '^N225', '^FCHI'
]

# Liste des unités de temps (ordonnée du plus petit au plus grand)
timeframes = ['30m', '60m', '90m', '1h', '1d', '5d', '1wk', '1mo', '3mo']

# Mapper les unités de temps avec des périodes compatibles
period_map = {
    '30m': '1mo',
    '60m': '1mo',
    '90m': '1mo',
    '1h': '1mo',
    '1d': '3mo',
    '5d': '6mo',
    '1wk': '1y',
    '1mo': '2y',
    '3mo': '5y'
}

# Fonction pour détecter les patterns englobantes et mesurer les mouvements et points culminants/plus bas
def detect_engulfing(data, long_term_horizon):
    signals = {'bullish_engulfing': [], 'bearish_engulfing': []}
    movements = {'bullish': [], 'bearish': []}
    long_term_movements = {'bullish': [], 'bearish': []}
    for i in range(1, len(data)):
        # Bullish Engulfing
        if data['Open'][i] < data['Close'][i-1] and data['Close'][i] > data['Open'][i-1]:
            signals['bullish_engulfing'].append(i)
            # Calculer le mouvement suivant l'englobante haussière jusqu'à un chandelier baissier
            for j in range(i+1, len(data)):
                if data['Close'][j] < data['Open'][j]:
                    movements['bullish'].append(data['Close'][j] - data['Close'][i])
                    break
            # Calculer le point culminant sur l'horizon de temps long
            if i + long_term_horizon < len(data):
                long_term_movements['bullish'].append(max(data['Close'][i:i+long_term_horizon]) - data['Close'][i])

        # Bearish Engulfing
        elif data['Open'][i] > data['Close'][i-1] and data['Close'][i] < data['Open'][i-1]:
            signals['bearish_engulfing'].append(i)
            # Calculer le mouvement suivant l'englobante baissière jusqu'à un chandelier haussier
            for j in range(i+1, len(data)):
                if data['Close'][j] > data['Open'][j]:
                    movements['bearish'].append(data['Close'][j] - data['Close'][i])
                    break
            # Calculer le point le plus bas sur l'horizon de temps long
            if i + long_term_horizon < len(data):
                long_term_movements['bearish'].append(min(data['Close'][i:i+long_term_horizon]) - data['Close'][i])
    return signals, movements, long_term_movements

# Détecter les patterns pour chaque actif et chaque unité de temps
engulfing_counts = {ticker: {tf: {'bullish': 0, 'bearish': 0, 'bullish_movements': [], 'bearish_movements': [], 'bullish_long_term': [], 'bearish_long_term': []} for tf in timeframes} for ticker in tickers}

long_term_horizon = 20  # 20 periods for long-term analysis (can be adjusted as needed)

for ticker in tickers:
    for tf in timeframes:
        try:
            period = period_map[tf]
            data = yf.download(ticker, period=period, interval=tf)
            if not data.empty:
                signals, movements, long_term_movements = detect_engulfing(data, long_term_horizon)
                engulfing_counts[ticker][tf]['bullish'] = len(signals['bullish_engulfing'])
                engulfing_counts[ticker][tf]['bearish'] = len(signals['bearish_engulfing'])
                engulfing_counts[ticker][tf]['bullish_movements'] = movements['bullish']
                engulfing_counts[ticker][tf]['bearish_movements'] = movements['bearish']
                engulfing_counts[ticker][tf]['bullish_long_term'] = long_term_movements['bullish']
                engulfing_counts[ticker][tf]['bearish_long_term'] = long_term_movements['bearish']
            else:
                print(f"No data available for {ticker} with interval {tf}")
        except Exception as e:
            print(f"Error downloading {ticker} with interval {tf}: {e}")

# Calculer la moyenne des mouvements pour chaque unité de temps et chaque actif
bullish_movements_mean_df = pd.DataFrame({tf: [np.mean(engulfing_counts[ticker][tf]['bullish_movements']) if engulfing_counts[ticker][tf]['bullish_movements'] else 0 for ticker in tickers] for tf in timeframes}, index=tickers)
bearish_movements_mean_df = pd.DataFrame({tf: [np.mean(engulfing_counts[ticker][tf]['bearish_movements']) if engulfing_counts[ticker][tf]['bearish_movements'] else 0 for ticker in tickers] for tf in timeframes}, index=tickers)

# Calculer la moyenne des points culminants/plus bas pour chaque unité de temps et chaque actif
bullish_long_term_mean_df = pd.DataFrame({tf: [np.mean(engulfing_counts[ticker][tf]['bullish_long_term']) if engulfing_counts[ticker][tf]['bullish_long_term'] else 0 for ticker in tickers] for tf in timeframes}, index=tickers)
bearish_long_term_mean_df = pd.DataFrame({tf: [np.mean(engulfing_counts[ticker][tf]['bearish_long_term']) if engulfing_counts[ticker][tf]['bearish_long_term'] else 0 for ticker in tickers] for tf in timeframes}, index=tickers)

# Calculer la fréquence des patterns englobantes
bullish_engulfing_freq_df = pd.DataFrame({tf: [engulfing_counts[ticker][tf]['bullish'] for ticker in tickers] for tf in timeframes}, index=tickers)
bearish_engulfing_freq_df = pd.DataFrame({tf: [engulfing_counts[ticker][tf]['bearish'] for ticker in tickers] for tf in timeframes}, index=tickers)

# Tracer les heatmaps de la fréquence des patterns englobantes
plt.figure(figsize=(14, 7))
sns.heatmap(bullish_engulfing_freq_df, annot=True, fmt='d', cmap='Blues', linewidths=0.5, linecolor='black')
plt.title('Frequency of Bullish Engulfing Patterns')
plt.xlabel('Timeframes')
plt.ylabel('Tickers')
plt.show()

plt.figure(figsize=(14, 7))
sns.heatmap(bearish_engulfing_freq_df, annot=True, fmt='d', cmap='Purples', linewidths=0.5, linecolor='black')
plt.title('Frequency of Bearish Engulfing Patterns')
plt.xlabel('Timeframes')
plt.ylabel('Tickers')
plt.show()

# Tracer les heatmaps des mouvements moyens après englobantes (pips)
plt.figure(figsize=(14, 7))
sns.heatmap(bullish_movements_mean_df.round(2), annot=True, fmt='.2f', cmap='Blues', linewidths=0.5, linecolor='black')
plt.title('Average Pips After Bullish Engulfing Patterns')
plt.xlabel('Timeframes')
plt.ylabel('Tickers')
plt.show()

plt.figure(figsize=(14, 7))
sns.heatmap(bearish_movements_mean_df.round(2), annot=True, fmt='.2f', cmap='Purples', linewidths=0.5, linecolor='black')
plt.title('Average Pips After Bearish Engulfing Patterns')
plt.xlabel('Timeframes')
plt.ylabel('Tickers')
plt.show()

# Tracer les heatmaps des points culminants/plus bas après englobantes (pips)
plt.figure(figsize=(14, 7))
sns.heatmap(bullish_long_term_mean_df.round(2), annot=True, fmt='.2f', cmap='Greens', linewidths=0.5, linecolor='black')
plt.title('Average Long-Term Peak Pips After Bullish Engulfing Patterns')
plt.xlabel('Timeframes')
plt.ylabel('Tickers')
plt.show()

plt.figure(figsize=(14, 7))
sns.heatmap(bearish_long_term_mean_df.round(2), annot=True, fmt='.2f', cmap='Oranges', linewidths=0.5, linecolor='black')
plt.title('Average Long-Term Lowest Pips After Bearish Engulfing Patterns')
plt.xlabel('Timeframes')
plt.ylabel('Tickers')
plt.show()

# Graphiques supplémentaires pour une analyse approfondie

# Histogramme de la fréquence des patterns englobantes par actif
for ticker in tickers:
    plt.figure(figsize=(10, 5))
    sns.barplot(x=timeframes, y=[engulfing_counts[ticker][tf]['bullish'] for tf in timeframes], palette='Blues')
    plt.title(f'Frequency of Bullish Engulfing Patterns for {ticker}')
    plt.xlabel('Timeframes')
    plt.ylabel('Frequency')
    plt.show()

    plt.figure(figsize=(10, 5))
    sns.barplot(x=timeframes, y=[engulfing_counts[ticker][tf]['bearish'] for tf in timeframes], palette='Purples')
    plt.title(f'Frequency of Bearish Engulfing Patterns for {ticker}')
    plt.xlabel('Timeframes')
    plt.ylabel('Frequency')
    plt.show()

# Boxplots des mouvements après englobantes pour visualiser la distribution
for tf in timeframes:
    plt.figure(figsize=(14, 7))
    sns.boxplot(data=[engulfing_counts[ticker][tf]['bullish_movements'] for ticker in tickers], palette='Blues')
    plt.title(f'Distribution of Bullish Engulfing Pattern Movements ({tf})')
    plt.xlabel('Tickers')
    plt.ylabel('Pips')
    plt.xticks(ticks=np.arange(len(tickers)), labels=tickers, rotation=90)
    plt.show()

    plt.figure(figsize=(14, 7))
    sns.boxplot(data=[engulfing_counts[ticker][tf]['bearish_movements'] for ticker in tickers], palette='Purples')
    plt.title(f'Distribution of Bearish Engulfing Pattern Movements ({tf})')
    plt.xlabel('Tickers')
    plt.ylabel('Pips')
    plt.xticks(ticks=np.arange(len(tickers)), labels=tickers, rotation=90)
    plt.show()

# Sélectionner les unités de temps les plus intéressantes en fonction de la fréquence des patterns et du nombre de pips moyens
best_timeframes = {}
for ticker in tickers:
    best_timeframes[ticker] = {}
    for tf in timeframes:
        freq_bullish = engulfing_counts[ticker][tf]['bullish']
        freq_bearish = engulfing_counts[ticker][tf]['bearish']
        avg_pips_bullish = np.mean(engulfing_counts[ticker][tf]['bullish_movements']) if engulfing_counts[ticker][tf]['bullish_movements'] else 0
        avg_pips_bearish = np.mean(engulfing_counts[ticker][tf]['bearish_movements']) if engulfing_counts[ticker][tf]['bearish_movements'] else 0

        score_bullish = freq_bullish * avg_pips_bullish
        score_bearish = freq_bearish * avg_pips_bearish

        best_timeframes[ticker][tf] = {'score_bullish': score_bullish, 'score_bearish': score_bearish}

# Afficher les unités de temps les plus intéressantes
for ticker in tickers:
    sorted_timeframes_bullish = sorted(best_timeframes[ticker].items(), key=lambda x: x[1]['score_bullish'], reverse=True)
    sorted_timeframes_bearish = sorted(best_timeframes[ticker].items(), key=lambda x: x[1]['score_bearish'], reverse=True)

    print(f"Best timeframes for {ticker} (Bullish):")
    for tf, score in sorted_timeframes_bullish:
        print(f"{tf}: {score['score_bullish']:.2f} pips")

    print(f"\nBest timeframes for {ticker} (Bearish):")
    for tf, score in sorted_timeframes_bearish:
        print(f"{tf}: {score['score_bearish']:.2f} pips")
    print("\n" + "-"*50 + "\n")

```
