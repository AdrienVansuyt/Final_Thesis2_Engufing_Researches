# **Corrélation / Décorrélation GOLD et PÉTROLE**

Matrice de corrélation avec heatmap sur les actifs que je trade pour faire de l'arbitrage

On test sur 20 ans

```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Liste des actifs à comparer
symbols = {
    'Gold': 'GC=F',
    'WTI Oil': 'CL=F',
    'NASDAQ': '^IXIC',
    'CAC 40': '^FCHI',
    'DAX': '^GDAXI',
    'Nikkei': '^N225',
    'S&P 500': '^GSPC',
    'Dow Jones': '^DJI',
    'Silver': 'SI=F',
    'Bitcoin': 'BTC-USD',
    'Ethereum': 'ETH-USD',
    'EUR/USD': 'EURUSD=X',
    'USD/JPY': 'JPY=X',
    'GBP/USD': 'GBPUSD=X',
    'USD/CHF': 'CHF=X',
    'USD/CAD': 'CAD=X'
}

# Téléchargement des données pour les actifs
data = yf.download(list(symbols.values()), start='2004-11-01', end='2024-11-01')['Adj Close']

# Renommer les colonnes pour les actifs
data.columns = list(symbols.keys())

# Filtrer pour ne garder que Gold et Oil en lignes
selected_assets = ['Gold', 'WTI Oil']
data = data[selected_assets + list(symbols.keys())[2:]]

# Calcul des rendements quotidiens
returns = data.pct_change()

# Calcul de la matrice de corrélation
corr_matrix = returns.corr().loc[selected_assets, symbols.keys()]

# Affichage de la matrice de corrélation avec seaborn
plt.figure(figsize=(14, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Matrice de corrélation pour Gold et Oil avec les autres actifs financiers et devises majeures")
plt.show()
```
